```python
import numpy as np 
import pandas as pd
from IPython.display import Markdown,display
def printmd(message,data):
    text="**<u>"+message+"</u>**"
    display(Markdown(text),data)

import matplotlib.pyplot as plt
```

```python
x_value_data=np.arange(0,5,0.1)
y_sin_value_data=np.sin(x_value_data)
y_cos_value_data=np.cos(x_value_data)
plt.plot(x_value_data, y_sin_value_data)
plt.title("sin()")
plt.xlabel('x (0-5, jumps 0.1)')
plt.ylabel('y')
plt.legend(['sin function'])
plt.show()
```

```python
fig = plt.figure(figsize=(6.4, 4.8))
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.set_xlabel('x (0-5, jumps 0.1)')
axes1.set_ylabel('y=sin(x)')
axes1.set_xlim(0,4.99)
axes1.set_ylim(-1.01,1.01)
axes1.set_xticks(np.arange(0, 5.1, 0.5)) 
axes1.plot(x_value_data,y_sin_value_data,label='sin(x)', color='blue')
axes1.legend(loc='upper right')

axes2 = fig.add_axes([1.2, 0, 1, 1])
axes2.plot(x_value_data, y_cos_value_data,label='cos(x)', color='red')
axes2.legend(loc='upper right')
axes2.set_xlabel('x (0-5, jumps 0.1)')
axes2.set_ylabel('y=cos(x)')
axes2.set_xlim(-0.02,5.01)
axes2.set_ylim(-1.01,1.02)
axes2.set_xticks(np.arange(0, 5.1, 0.5)) 
plt.show()
```
