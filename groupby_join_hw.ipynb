```python
import numpy as np
import pandas as pd
from IPython.display import Markdown,display
def printmd(message,data):
    text="**<u>"+message+"</u>**"
    display(Markdown(text),data)
```

```python
products_df = pd.DataFrame({
'Product_ID': [1, 2, 3],
'Product_Name': ['iPhone 12', 'Samsung S21', 'Pixel 5'],
'Product_Price': [999, 899, 699],
})

sales_df = pd.DataFrame({
'Sale_ID': [1, 2, 3, 4, 5, 6, 7],
'Sale_Date': pd.to_datetime(['2021-01-01', '2021-02-01', '2021-02-15', '2020-03-01', '2020-04-01',

'2020-02-15', '2020-01-01']),
'Customer_ID': [1, 2, 3, 2, 3, 4, 4],
'Product_ID': [1, 1, 2, 2, 3, 3, 1],
})

customers_df = pd.DataFrame({
'Customer_ID': [1, 2, 3, 4],
'Customer_Name': ['John Doe', 'Jane Doe', 'Jim Brown', 'Jake Smith'],
'Customer_Age': [25, 30, 35, 20],
'Customer_Email': ['john@doe.com', 'jane@doe.com', 'jim@brown.com', 'jake@smith.com'],
})
merge_df=pd.merge(products_df,sales_df,on='Product_ID')
merge_df=pd.merge(merge_df,customers_df, on='Customer_ID')
merge_df
```

```python
#1 For each product calculate the number of customers that bought it and their average age
products_groups=merge_df.groupby('Product_Name')
counts=products_groups.size().reset_index(name='buyers')
avg_ages=products_groups['Customer_Age'].mean().reset_index(name='Avg Customer_Age')
results=pd.merge(counts,avg_ages,on='Product_Name')
printmd('number of customers and their average age for each product:',results)
```

```python
#For each year calculate the total sales for each product for that year
merge_df['year']=merge_df['Sale_Date'].dt.year
year_product_groups=merge_df.groupby(['Product_Name','year'])
counts=year_product_groups.size().reset_index(name='count')
result=pd.merge(counts,products_df,on='Product_Name')
result['total sale']=result['count']*result['Product_Price']
printmd('total sale for each product for each year:',result)
```
